### Fill missing value with mean

```python
df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean(),inplace = True)
```

```python
df.isna().sum()
```

```python
categorical_columns = ['Gender', 'Customer Type', 'Type of Travel','Class', 'satisfaction']

for i in categorical_columns: 
    print(f'Unique values in {i} are------>{df[i].unique()}')


```

```python
def satisfied(x):
    if x == 'neutral or dissatisfied':
        return 0
    else:
        return 1
```

```python
df['satisfaction'] = df['satisfaction'].apply(satisfied)
```

```python
df['satisfaction'].unique()
```

```python
df.head()
```

```python
df['satisfaction'].unique()
```

```python
df.drop(columns = ['Unnamed: 0','id'],inplace = True)
```

### Change the rest of the categorical data into nominal using OneHotEncoding

```python
pd.get_dummies(df.drop(columns = ['satisfaction']), drop_first = True)
```

### Plot a pairplot of the dataset

```python
df.head()
```

```python
df = pd.get_dummies(df,drop_first=True)
df.head()
```

```python
X = df.drop(columns = ['satisfaction'])
y = df['satisfaction']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X_scaled,columns = X.columns)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Create Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test , y_pred)
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(metrics.confusion_matrix(y_test, y_pred),annot = True,fmt = 'd')
```

```python
print(metrics.classification_report(y_test,y_pred))
```
